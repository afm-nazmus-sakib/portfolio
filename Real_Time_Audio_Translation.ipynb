# 2. Audio → Text (STT)
Use OpenAI’s Whisper to transcribe the audio to English text

```python
import whisper

model = whisper.load_model("base")
result = model.transcribe("/content/drive/MyDrive/Colab Notebooks/Real-Time Audio Translation/InputAudio.m4a")
english_text = result["text"]
print("Transcription:", english_text)
```

# 3. Translate (English → German)
Translate the text into German using a Huggingface model – specifically MarianMT

```python
from transformers import MarianMTModel, MarianTokenizer

src_text = english_text
tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-de")
model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-de")

translated = model.generate(**tokenizer(src_text, return_tensors="pt", padding=True))
german_text = tokenizer.decode(translated[0], skip_special_tokens=True)
print("Translation:", german_text)
```

# 4. Text → Speech (TTS)

```python
!apt-get install espeak
```

use Coqui TTS to generate natural-sounding German audio from the translated text

```python
from TTS.api import TTS

tts = TTS(model_name="tts_models/de/thorsten/tacotron2-DDC", progress_bar=True)
tts.tts_to_file(text=german_text, file_path="/content/drive/MyDrive/Colab Notebooks/Real-Time Audio Translation/OutputAudio.m4a")
```

# 5. Wrap in a Streamlit App


```python
import streamlit as st

st.title("Audio Translator")

uploaded_file = st.file_uploader("Upload English Audio", type=["m4a"])

if uploaded_file:
    # Save and transcribe
    with open("temp.m4a", "wb") as f:
        f.write(uploaded_file.read())
    result = model.transcribe("temp.m4a")
    st.write("English:", result["text"])

    # Translate
    translated = model.generate(**tokenizer(result["text"], return_tensors="pt", padding=True))
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    st.write("German:", translated_text)

    # TTS
    tts.tts_to_file(text=translated_text, file_path="out.m4a")
    st.audio("out.m4a")
```
